```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)

y = np.sin(x)

plt.plot(x, y)

plt.title("Graph of Sine Function")
plt.xlabel("x values")
plt.ylabel("sin(x)")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)

y = np.cos(x)

plt.plot(x, y)

plt.title("Graph of Cosine Function")
plt.xlabel("x values")
plt.ylabel("cos(x)")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

y = np.tan(x)

plt.plot(x, y)

plt.title("Graph of Tangent Function")
plt.xlabel("x values")
plt.ylabel("tan(x)")

plt.ylim(-10, 10)

plt.show()
```

```python
import numpy as np

s1 = np.array(["apple", "banana", "cherry"])
s2 = np.array(["juice", "shake", "pie"])

c = np.char.add(s1, s2)
u = np.char.upper(s1)
l = np.char.lower(s1)
t = np.char.title(s1)
r = np.char.replace(s1, "a", "@")

print("Concatenation:", c)
print("Uppercase:", u)
print("Lowercase:", l)
print("Title case:", t)
print("Replace operation:", r)
```

```python
import pandas as pd
import numpy as np
d = {"Name": ["Aman", "Riya", "Karan", "Neha"],
     "Marks": [85, np.nan, 78, np.nan]}
df = pd.DataFrame(d)
print("Original Data:")
print(df)
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
print("After handling missing values:")
print(df)
```

```python
import pandas as pd
d1 = {"Name": ["Aman", "Riya"], "Marks": [85, 90]}
d2 = {"Name": ["Karan", "Neha"], "Marks": [78, 88]}
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
df3 = pd.concat([df1, df2])
print("Concatenated DataFrame:")
print(df3)
```

```python
import pandas as pd

d1 = {"Name": ["Aman", "Riya", "Karan"], "Marks": [85, 90, 78]}
d2 = {"Grade": ["A", "A+", "B+"]}

df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)

df3 = df1.join(df2)

print("Joined DataFrame:")
print(df3)
```

```python
import pandas as pd

d1 = {"Name": ["Aman", "Riya"], "Marks": [85, 90]}
d2 = {"Name": ["Karan"], "Marks": [78]}

df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)

df3 = pd.concat([df1, df2], ignore_index=True)

print("Data after append:")
print(df3)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

d = {
    "Day": [1, 2, 3, 4, 5],
    "Sales_A": [10, 15, 13, 17, 20],
    "Sales_B": [8, 12, 11, 15, 18]
}

df = pd.DataFrame(d)

sns.lineplot(x="Day", y="Sales_A", data=df, label="Sales A")
sns.lineplot(x="Day", y="Sales_B", data=df, label="Sales B")

plt.title("Multiple Line Plot")
plt.show()
```
